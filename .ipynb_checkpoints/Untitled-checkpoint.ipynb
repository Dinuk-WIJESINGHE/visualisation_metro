```python
import folium
from folium.plugins import MarkerCluster
```

```python
df_geo = df_clean_test_all_lines[["Geo Point","Geo Shape","nom_long","Name_Line"]]
print(df_geo)
```

```python
df_geo = df_geo.copy()
df_geo[['latitude', 'longitude']] = df_geo['Geo Point'].str.split(',', expand=True)

print(df_geo)
```

```python
print(df_geo['latitude'].astype(float).mean());
```

```python
# Create a base map
m = folium.Map(location=[df_geo['latitude'].astype(float).mean(), df_geo['longitude'].astype(float).mean()], zoom_start=4)

# Add markers
for _, row in df_geo.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['nom_long']
    ).add_to(m)

# Display the map
m
```
